# Movimiento con resistencia del aire

```python
import numpy as np

class Movimiento_friccion:

    def __init__(self, a, b):

        self.a, self.b = a, b

    def __call__(self, u, t):

        v, a, b = u, self.a, self.b 

        return a - b*v
```

```python
import plotly.graph_objects as go
import sys
sys.path.append("../")
from PhysicsPy.ODEsMethods import *	

Sistema = Movimiento_friccion(10, 1)

Solucion = Euler(Sistema)
Solucion.InitialConditions(0, [0,10], 0.01)
v_x,t = Solucion.SolveODE()

v_x_sol = Sistema.a - (Sistema.a)*np.exp(-t)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=t, y=v_x[:,0], name='Solución numérica',
                         line=dict(color='purple', width=3)))
fig.add_trace(go.Scatter(x=t, y=v_x_sol, name = 'Solución analítica',
                         line=dict(color='pink', width=2)))

fig.update_layout(title='Movimiento en resistencia del aire',
                   xaxis_title='tiempo (s)',
                   yaxis_title='velocidad (m/s)')

fig.show()
```
